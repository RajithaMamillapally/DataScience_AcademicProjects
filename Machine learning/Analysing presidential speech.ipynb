In this particular project, we are going to work on the inaugural corpora from the nltk in Python. We will be looking at the following speeches of the Presidents of the United States of America:

3.	President Richard Nixon in 1973
2.1) Find the number of characters, words and sentences for the mentioned documents. (Hint: use .words(), .raw(), .sent() for extracting counts)

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
import nltk
nltk.download('inaugural')
from nltk.corpus import inaugural
inaugural.fileids()
inaugural.raw('1941-Roosevelt.txt')
inaugural.raw('1961-Kennedy.txt')
inaugural.raw('1973-Nixon.txt')
nltk.download("stopwords")
nltk.download("punkt")
```

```python
print("Number of characters in 1941-Roosevelt.txt are {}". format(len(inaugural.raw(('1941-Roosevelt.txt')))))
print("Number of words in 1941-Roosevelt.txt are {}". format(len(inaugural.words(('1941-Roosevelt.txt')))))
print("Number of sentences in 1941-Roosevelt.txt are {}". format(len(inaugural.sents(('1941-Roosevelt.txt')))))
```

```python
print("Number of characters in 1961-Kennedy.txt are {}". format(len(inaugural.raw(('1961-Kennedy.txt')))))
print("Number of words in 1961-Kennedy.txt are {}". format(len(inaugural.words(('1961-Kennedy.txt')))))
print("Number of sentences in 1961-Kennedy.txt are {}". format(len(inaugural.sents(('1961-Kennedy.txt')))))
```

```python
print("Number of characters in 1973-Nixon.txt are {}". format(len(inaugural.raw(('1973-Nixon.txt')))))
print("Number of words in 1973-Nixon.txt are {}". format(len(inaugural.words(('1973-Nixon.txt')))))
print("Number of sentences in 1973-Nixon.txt are {}". format(len(inaugural.sents(('1973-Nixon.txt')))))
```

#2.2) Remove all the stopwords from the three speeches. Show the word count before and after the removal of stopwords. Show a sample sentence after the removal of stopwords.

```python
import string
stopwords = nltk.corpus.stopwords.words('english') +list(string.punctuation)
stopwords.extend(['day','since','1789', 'task', 'within', 'us', '--','way','seen','still','yet','came','go','also','mr','l','let','know','life','new','ago','shall','always','...','.','."','may','sides','mrs','one','people','men','seem','full','something','upon','man','away','wholly','alone','year','human','create','ten','free','common','less','right','eight','even','every','time','ask','alike','days','end','dare','call','first','better','others','turning','often','greatly','clearly','role','good','able'])
```

```python
print(stopwords)
```

#Roosevelt

```python
all_words=[]
def lower(x):
  for word in inaugural.words("1941-Roosevelt.txt"):
    words=word.lower()
    all_words.append(words)
lower(inaugural.words(("1941-Roosevelt.txt")))
all_words
```

```python
clean_words=[]
def clean(x):
  for word in x:
    if word not in stopwords:
      clean_words.append(word)
clean(all_words)
clean_words
```

```python
print("Number of words in 1941-Roosevelt.txt after cleaning stop words are {}". format(len(clean_words)))
```

```python
Roosevelt_wordcloud=' '.join(clean_words)
Roosevelt_wordcloud
```

#2.4) Plot the word cloud of each of the three speeches. (after removing the stopwords)

```python
Roosevelt_wordcloud=' '.join(clean_words)
wordcloud = WordCloud(width = 2000, height =2000,
                background_color ='black',
                min_font_size = 10, random_state=100).generate(Roosevelt_wordcloud)

plt.figure(figsize = (6, 6), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.xlabel('Word Cloud')
plt.tight_layout(pad = 0)

print("Word Cloud for Speech of Roosevelt!!")
plt.show()
```
